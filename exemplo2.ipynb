Aprendendo Python

Comandos para instalar venv.
source .venv/bin/activate

Depois de ativar, usar: pip list, para ver a lista de bibliotecas.
Depois atualizar o pip: 
python3 -m pip install -- upgrade pip

Instalando a biblioteca Pandas
pip install pandas numpy matplotlib
Pandas = a mais usada, organiza os dados em tabelas
Numpy = [1234-faz matematica de alto nível] a base de tudo
Matplotlib = (o desenhista, mostra os dados em gráficos)


Depois de instalar, podemos executar:
pip list: para ver a lista de biblioteca instalada
pip freeze: para ver as versões de cada biblioteca instalada

Para salvar as versões das bibliotecas
pip freeze > requirements.txt
Para adicionar uma nova bibliooteca:
Abrir requirements.txt e digitar lá a biblioteca (exemplo: seaborn)
Depois execultar:
pip install -r requirements.txt (para instalar a nova biblioteca seaborn)
E depois dar um :
pip freeze > requirements.txt
